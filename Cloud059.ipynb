```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

plt.style.use('bmh')
df = pd.read_csv('/prices.csv')

# All Brands
x = df['Model']
y = df['Price']

# Specified Brands
# specified = df.loc[df['Company'] == 'Samsung']
# x = specified['Model']
# y = specified['Price']



# Pie chart
plt.pie(y, labels=x, radius=1.2,autopct='%0.01f%%', shadow=True, explode=[.05,.2,.05,.2,.05,.2,.05])


plt.show()
```

```python
# Line Graph
plt.xlabel('Model', fontsize=18)
plt.ylabel('Price($)', fontsize=16)
plt.scatter(x, y)
plt.plot(x, y)
```

```python
# Bar chart
plt.xlabel('Model', fontsize=10)
plt.ylabel('Price($)', fontsize=10)
plt.bar(x, y)
plt.xticks(rotation = 90)
```

```python
sns.set(font_scale=1.4)
df['Price'].plot(kind='hist', figsize=(10, 10));
plt.xlabel("Price", labelpad=14)
plt.ylabel("frequency", labelpad=14)
plt.title("histogram", y=1.015, fontsize=17);
```
